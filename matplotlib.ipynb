# bar chart

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = np.array([1,2,3,4,5])
Data = np.array([1000,2000,3500,4500,5000])
AI = np.array([1500,2500,3500,5500,6000])
```

```python
#%matplotlib ipympl
plt.bar(age,AI , alpha = 0.7 , label = 'AI',color = 'blue' )   #alpha is transparency
plt.bar(age,Data , label = 'data', color = 'green')
plt.legend()
plt.show()
```

```python
plt.barh(age,AI , alpha = 0.7 , label = 'AI',color = 'blue' )   #alpha is transparency
plt.barh(age,Data , label = 'data', color = 'green')
plt.legend()
plt.show()
```

# pie chart

```python
plt.pie([100,80,50] , labels= ['sam' , 'shia' , 'apple'], autopct= '%.2f', colors= ['b', 'r','g'],
       startangle= 90 , explode=(0,0,0.1), shadow=True )
plt.show()
```

# subplot Figure

show several chart

```python
fig = plt.figure()
```

```python
x = np.linspace(0,2 * np.pi,100)
```

```python
fig = plt.figure(figsize= (8,8))
plt1 = fig.add_subplot(221)
plt2 = fig.add_subplot(222)
plt3 = fig.add_subplot(223)
plt4 = fig.add_subplot(224)

plt1.plot(x, np.sin(x), label ='sin')
plt1.set_title('sin')

plt2.plot(x, np.cos(x), label = 'cos')
plt3.scatter(x[:10], np.sin(x)[:10])
plt4.plot(x, np.tan(x))

plt1.legend()
plt2.legend()

fig.subplots_adjust(hspace = 0.5 , wspace = 0.5)
plt.show()
```

```python
fig2,(plt1,plt2) = plt.subplots(1,2)  #1 row 2 column
plt1.plot(x, np.sin(x), color ='red')
plt2.plot(x, np.cos(x))

fig2.subplots_adjust( wspace = 0.5)

plt.show()
```
