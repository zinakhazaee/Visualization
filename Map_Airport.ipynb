```python
import pandas as pd
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('airports.csv')
df.head(5)
```

```python
fig = px.scatter_geo(df[df['type']=='large_airport'],lat='latitude_deg',lon='longitude_deg', hover_name='name')
fig.update_layout(title = 'Airports', title_x=0.5)
fig.show()
```
